```python
# Importing necessary library
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Reading the data
data = pd.read_csv("winequality-white.csv",sep=';',quotechar='"')
print(data.keys(), end='\n\n')
print(data.shape, end='\n\n')
n = data.shape[0]
data.head()
```

```python
# Replace 0 with min
for feature in data.columns.tolist():
    mini = 1000
    for row in range(n):
        if data[feature][row] > 0 and data[feature][row] < mini:
            mini = data[feature][row]
    data[feature] = data[feature].replace(0, mini)

data.head()
```

```python
# log transformation
log_data = np.log(data)
log_data = log_data.drop('quality', axis = 1)
log_data = pd.concat([log_data, data['quality']], axis = 1, copy = False)
log_data
```

```python
# Histogram of different features in the dataset
for feature in data.columns.tolist():
    print("Feature: ", feature)
    # the histogram of the data
    n, bins, patches = plt.hist(log_data[feature])
    plt.show()
```

# OUTLIERS DETECTION

### Using interquartile range to detect outliers

```python
outliers = set()

for feature in log_data.columns.tolist():
    print("Feature:", feature)
    
    # Calculate Q1
    Q1 = np.percentile(log_data[feature], 25.)
    
    # Calculate Q3
    Q3 = np.percentile(log_data[feature], 75.)
    
    # Calculate IQR
    step = (Q3 - Q1)*1.5
    print("Step for outliers detection:", str(step))
    
    # Identify and display the outliers
    feature_outliers = log_data[~((log_data[feature] >= Q1 - step) & (log_data[feature] <= Q3 + step))]
    #display(feature_outliers)
    print("The number of outliers detected: ", str(len(feature_outliers)), '\n\n')
    
    for data_points in feature_outliers.index.tolist():
        outliers.add(data_points)
    
iqr_data = log_data.drop((log_data.index[list(outliers)])).reset_index(drop=True)
print("The new data has {} samples with {} features each.".format(iqr_data.shape[0], iqr_data.shape[1]))
```

### Median-Absolute-Value (MAD) to detect outliers

```python
def mad_based_outlier(points, thresh=3.5):
    if len(points.shape) == 1:
        points = points[:,None]
    median = np.median(points, axis=0)
    diff = np.sum((points - median)**2, axis=-1)
    diff = np.sqrt(diff)
    med_abs_deviation = np.median(diff)

    modified_z_score = 0.6745 * diff / med_abs_deviation

    return modified_z_score > thresh

mad_outliers=set()

for feature in log_data.columns.tolist():
    print("Feature:", feature)
    
    # Identify and display the outliers
    feature_outliers = log_data[feature][mad_based_outlier(data[feature])==True]
    
    #display(feature_outliers)
    print("The number of outliers detected: ", str(len(feature_outliers)), '\n\n')
    
    for data_points in feature_outliers.index.tolist():
        mad_outliers.add(data_points)
        
mad_data = log_data.drop((log_data.index[list(mad_outliers)])).reset_index(drop=True)
print("The new data has {} samples with {} features each.".format(mad_data.shape[0], mad_data.shape[1]))
```

```python
# Histogram of different features in the dataset
for feature in mad_data.columns.tolist():
    print("Feature: ", feature)
    # the histogram of the data
    n, bins, patches = plt.hist(mad_data[feature])
    plt.show()
```

The MAD method was a more appropriate method for outliers detection since it discard less points in a large dataset compared to the IQR method.

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(mad_data.drop('quality', axis = 1), mad_data['quality'], test_size=0.2, random_state = 22)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

# FEATURE SELECTION

### Principal Component Analysis

```python
from sklearn.decomposition import PCA
pca = PCA(7)
pca.fit(X_train)
X_train = pca.transform(X_train)
X_test = pca.transform(X_test)

pca_features = range(pca.n_components_)

plt.bar(pca_features, pca.explained_variance_)
plt.xticks(pca_features)
plt.show()
```

```python
# Fitting the model

from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.metrics import accuracy_score

# Logistic Regression
lr = LogisticRegression(C = 1)
lr.fit(X_train, y_train)
print("Logistic Regression: ")
print("R^2 Score: ", lr.score(X_test, y_test))
print("Root Mean Squared Error: ", np.sqrt(lr.score(X_test, y_test)))
```

```python
# Random Forest Classifier
dt = RandomForestClassifier(max_depth = 41, n_estimators = 100)
dt.fit(X_train, y_train)
print("Random Forest: ")
print("R^2 Score: ", dt.score(X_test, y_test))
print("Root Mean Squared Error: ", np.sqrt(dt.score(X_test, y_test)))
```

```python
#SVC
svc = SVC(C = 10, gamma = 1, kernel = 'rbf')
svc.fit(X_train, y_train)
print("R^2 Score: ", svc.score(X_test, y_test))
print("Root Mean Squared Error: ", np.sqrt(svc.score(X_test, y_test)))
```
